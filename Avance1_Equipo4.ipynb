Utilizamos el método de interpolación lineal para interpolar los valores faltantes del dataset debido a que es una técnica comúnmente utilizada y de facil implementacion para estimar valores intermedios dentro de un conjunto de datos, especialmente cuando se trata de datos secuenciales o tabulares. La justificación detrás de este método radica en la suposición de que los datos tienen un comportamiento lineal entre puntos conocidos.

Ademas la interpolación lineal en Pandas se integra bien con otras funcionalidades de la biblioteca. Por ejemplo, después de la interpolación, es posible aplicar otras operaciones de Pandas como filtrado, agrupación o cálculos adicionales. Asi como ofrece flexibilidad en la forma en que se puede aplicar la interpolación lineal. Podemos especificar el método de interpolación (**linear**, **time**, **index**) asi como la dirección (**forward**, **backward**, **both**, **nearest**) según las necesidades específicas del analisis.

Por ultimo la interpolación lineal se puede aplicar a diferentes tipos de datos presentes en un DataFrame de Pandas. Todos los puntos mencionados anteriormente justifican por que este metodo es una opción sólida para el análisis de datos en Python.





```python
df.interpolate(method ='linear', limit_direction ='forward')
```
