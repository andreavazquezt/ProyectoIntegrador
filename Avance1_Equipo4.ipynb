Utilizamos el mtodo de interpolacin lineal para interpolar los valores faltantes del dataset debido a que es una tcnica comnmente utilizada y de facil implementacion para estimar valores intermedios dentro de un conjunto de datos, especialmente cuando se trata de datos secuenciales o tabulares. Por ejemplo, despus de la interpolacin, es posible aplicar otras operaciones de Pandas como filtrado, agrupacin o clculos adicionales. Asi como ofrece flexibilidad en la forma en que se puede aplicar la interpolacin lineal. Podemos especificar el mtodo de interpolacin (**linear**, **time**, **index**) asi como la direccin (**forward**, **backward**, **both**, **nearest**) segn las necesidades especficas del analisis.

Por ultimo la interpolacin lineal se puede aplicar a diferentes tipos de datos presentes en un DataFrame de Pandas. Todos los puntos mencionados anteriormente justifican por que este metodo es una opcin slida para el anlisis de datos en Python.
