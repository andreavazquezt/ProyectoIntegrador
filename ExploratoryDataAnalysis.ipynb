El proposito del siguiente notebook es ejecutar herramientas basicas utilizadas dentro del ***Exploratory Data Analysis*** para comenzar a familiarizarnos con las estrategias que nos ayudaran a limpiar y explorar nuestros datos, cuando terminemos de recolectarlos.

Comenzamos importando las librebrias que utilizaremos para el desarrollo del codigo:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Cargamos un data set de prueba para practicar con las herramientas de ***EDA***:

```python
df = pd.read_csv("/content/drive/MyDrive/world_population.csv")
df
```

Visualizamos la informacion a nivel general del dataset, como el numero de columnas, el nombre, los valores y el tipo de datos:

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

Visualizamos la informacion por columnas con el siguiente comando:

```python
df.describe()
```

Para poder descartar los valores faltantes y evitar que cambien nuestro conjunto de datos, utilizamos el siguiente comando para ubicarlos y asi poder ejecutar la estrategia correspondiente y ajustar nuestro dataset. Lo cual es especificamente tratamos de hacer durante el ***Exploratory Data Analysis***, encontrar valores faltantes, valores atipicos en otras palabras, cosas que esten mal con nuestro dataset o inclusive encontrar diversas perspectivas dentro de la misma busqueda:

```python
df.isnull().sum()
```

Para poder visualizar los valores unicos de cada columna utilizaremos el siguiente comando:

```python
df.nunique()
```

Para poder obtener las columnas donde los tipos de datos que incluyan un valor numerico:

```python
df.select_dtypes(include='number')
```

En caso de requerirlo tambien podemos filtrarlo por objeto, en resumen el comando nos permite especificar el tipo de datos y filtrarlo por la columna de interes:

```python
df.select_dtypes(include='object')
```

Para ordenar de manera ascendente nuestros datos podemos utilizar el siguiente comando, para efectos del ejemplo demostraremos solo para las primeras 5 filas del dataset y para la columna de ***2022 Population***:

```python
df.sort_values(by="2022 Population").head()
```

En caso de ser necesario ordenar los datos de manera descendente y para un numero superior de filas, podemos cambiarlo de la siguiente manera:

```python
df.sort_values(by="2022 Population", ascending=False).head(10)
```

Con el objetivo de explorar las relaciones entre las variables importantes de nuestro dataset efectuamos el siguiente analisis de correlacion cuantitativo y grafico:

```python
df.corr(numeric_only=True)
```

Tambien generamos nuestro heatmap, ajustando la escala de la imagen para tener una mejor visualizacion que nos ayude en el analisis de nuestro dataset:

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

Tambien podemos imprimir las columnas de nuestro dataset en caso de requerirlo para el analisis de nuestros datos:

```python
df.columns
```

Con la siguiente funcion podemos agrupar nuestros datos por continente o cualquier otra categoria de interes, especificando las columnas a ordenar en este ejemplo de forma descendente pero por default la funcion lo ordena en forma ascendente:

```python
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']].mean().sort_values(by='2022 Population', ascending=False)
df2
```

Si necesitamos filtrar nuestro dataset por alguna ***categoria/columna*** especifica podemos utilizar el siguiente comando:

```python
df2 = df[df['Continent'].str.contains('Oceania')]
df2
```

Para intercambiar las columnas por las filas podemos utilizar el siguiente comando en caso de requerirlo:

```python
df2.transpose()
```

Tambien podemos generar un grafico de nuestro dataset, para visualizar nuestros datos de manera mas eficiente:

```python
df.plot()
```

Para encontrar los ***valores atipicos ***de nuestro dataset podemos utilizar la siguiente herramienta lo cual nos ayudara a visualizarlos con mayor facilidad, ajustando la escala de la grafica segun corresponda:  

```python
df.boxplot(figsize=(20,10))
```
